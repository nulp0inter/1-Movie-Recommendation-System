```python
'''
column of movies = 20, credits = 4 and after merging on basis of 'title' column = 23 as 'title' column willnot be 
remerged. Now as we have to create a movie recommendation system we have to create a tag for each movie, and for
which we will only keep columns which will help us to create tags.

For example 'budget' might not be important. If a person likes high budget movie then its not highly probable that 
the same person may like another high budget movie. But if a person likes horror movies than it os likely that that
person may like another horror movie.

So, helpful columns are, 'genres', 'id' (important for identifying movies uniquely), 'keywords', 'title','overview',
'cast', 'crew'
'''

movies.info()
```

```python
# now we will further merge 'overview','genres','keywords','cast', 'crew' and create 3 columns

'''
Now we will change the format of genras, keyword, cast (we will merge only top 3 crews), crew (director only) and 
merge 

First check for missing data, duplicate data and if soo we will replace it

'''

movies.isnull().sum()
```

```python
'''
Here we have to chnage list of dictionary to list ie.
['Action','Adventure','Fantasy','Science Fiction']

here ast.literal_eval() convert string to list
'''

import ast

def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies
```

```python
'''
for cast we will only put top 3 cast and only director in crew 
'''

movies['cast'][0]
```

```python
"Here we will only take 3 dictionary"

def convert_cast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if (counter!=3):
            L.append(i['name'])
            counter = counter + 1
        else:
            break
    return L  
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies['cast']
```

```python
movies['crew'][0][:10]
```

```python
''' We will only need director information so '''

def convert_crew(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if (i['job']=='Director'):
            L.append(i['name'])
            break   
    return L  
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)

movies['crew']
```

```python
movies.head(1)
```

```python
'''
The overview is a string so we will split it to individual strings i.e In, the, 22nd, ...
'''

movies['overview'] = movies['overview'].apply(lambda x:x.split())

movies.head(1)
```

```python
'''
Before merging what we have to do is remove space in cast and crew because while creating tags we may encounter
problems of same initial name but different last name 'sam worthington' and 'sam mandes' and this will mess our 
recommendation system. If a person is interested in 'sam mandes' movies system might recommend 'sam wortgington' 
movies if space is not removed. 
'''

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
''' creating a tag column in movies '''

movies['tag'] = movies['overview']+  movies['genres']+ movies['keywords']+ movies['cast']+ movies['crew']

movies.head()
```

```python
#now creating a new data frame with only necessary columns

movies = movies.drop(columns=['budget',
'genres',
'homepage',
'id',                
'keywords',                    
'original_language',           
'original_title',              
'overview',                    
'popularity',                  
'production_companies',        
'production_countries',        
'release_date',                
'revenue',                     
'runtime',                     
'spoken_languages',            
'status',                      
'tagline',                                        
'vote_average',                
'vote_count',                                    
'cast',                        
'crew',])

movies
```

```python
#removing spcae

movies['tag'] = movies['tag'].apply(lambda x: " ".join(x))
movies
```

```python
movies['tag'] = movies['tag'].apply(lambda x:x.lower())
movies
```

```python
# DATA MINING PART #
'''
Now from all tags we will combine them
  tag = tag_1 + tag_2 + tag_3 +...+ tag_n

now from this combined tag we will remove stop words i,e. are, a , in, to, the, ... (word that help to create 
sentence but do not have meaning in itself). Then we will generate 5000 frequent words, and then we will generate a 
frequency table of each tag with 5000 words. This is text vectorization method and other techniques can also be uesd 
suppose vectors are

[0 0 2 3 4 5 0 0]
[0 1 3 2 1 0 3 1]

then we will measure similarity between these vector. Based upon closest similarity we will recommned movie.
''' 

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(movies['tag']).toarray()
vector
```

```python
for i in range(100):
    print(cv.get_feature_names_out()[i], end =", ");
```

```python
'''
Now if you look at the data then their are two problems here

1. First problem is words like
     abilities, ability
     action, actions
These words should be made same

2. Numerical values
    00, 000, 007, 10, 100, 11, 12, 14, 15, 16, 17
    Howerver alphanumeric values like 60s, 1930s is not a problem
    
To solve this problem we can apply stemming a NLP module. 
     After stemming ['love','loved','loves']=['love','love,'love']
                    ['dancing','dance','danced']=['dance','dance','dance']

'''
```

```python
#for steaming

import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer() 
```

```python
#splitting the text and applying stem 

def stem(text):
    k = []
    for j in text:
        info = ""
        for i in j.split():
            i = ps.stem(i)
            info = info + " " + i
        k.append(info)  
    return k
```

```python
#Demo of how stem works in list of string
text = ["i loved watching my football team play every weekend.","i also enjoyed nba."]
print(stem(text))
```

```python
print(movies['tag'])
```

```python
#steming

movies['tag'] = stem(movies['tag'])
print(movies['tag'])
```

```python
# recreating vectors again after steming

vector = cv.fit_transform(movies['tag']).toarray()
```

```python
'''
Now from these vectors we will calculate a cosine similarity between other movies and based ot the cosine similarity we will 
recommend movies. Here we have 5000 movies = 5000 dimension, and as dimention increases curse of dimensionality also increases.

similarity = 1/distance

more similarity less distance and vice versa
'''

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector) #calculate a cosine similarity of a vector with every ither vector
similarity                             #here diagonal is equal to 1 because similarity withitself will be maximum
```

```python
#now the function that returns 5 similar movies based on similarity index

def recommend_list(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    '''
    Now if we print(movie_list) this movie list we get 
    [(579, 0.26549712202679576), (331, 0.23097828906119441), (423, 0.2182178902359924), (51, 0.21602468994692867), 
    (118, 0.2020305089104422)]
    so we will use movie_id and find movies names

    movie_list[0]=579,movie_list[0]=331,...
    '''
    recommend_movie=[]
    for i in movies_list:
        recommend_movie.append(movies.iloc[i[0]].title)
    return recommend_movie 
```
